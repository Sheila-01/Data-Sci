```python
import matplotlib.pyplot as plt

plt.scatter(data['Total Revenue'], data['Total Cost'])
plt.xlabel('Total Revenue', fontsize = 15)
plt.ylabel('Total Cost', fontsize=15)
plt.xticks([0,1000000, 2000000, 3000000,4000000,5000000,6000000, 7000000], ['0','1M','2M','3M','4M','5M','6M','7M'])
plt.yticks([0,1000000, 2000000, 3000000,4000000,5000000], ['0','1M','2M','3M','4M','5M'])
plt.title('Sales Records', fontsize = 15)

plt.show()
```
